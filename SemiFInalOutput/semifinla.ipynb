### Data Processsing ###

```python
data.drop(["id", "language","votes_up", "comment_count", "steam_purchase", "recieved_for_free", "written_during_early_access", "written_during_early_access", "author_num_reviews" ], axis =1)

print(data.columns)
```

```python
from sklearn.model_selection import train_test_split

x = data["review"]
y = data["voted_up"]

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words= "english", max_df=0.7)

X_train_tfidf = vectorizer.fit_transform(X_train.values.astype('U'))

X_test_tfidf = vectorizer.transform(X_test.values.astype('U'))
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(X_train_tfidf, y_train)
```

### Evaluate the Model ###

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred = model.predict(X_test_tfidf)

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.4f}')

print(classification_report(y_test, y_pred))
```

```python
new_reviews_tfidf = vectorizer.transform(data["review"].values.astype('U'))

predictions = model.predict(new_reviews_tfidf)

for review, label in zip(data["review"].values.astype('U'), predictions):
    sentiment = 'Positive' if label == 1 else 'Negative'
    print(f'Review: {review} | Sentiment: {sentiment}')
```
